Text Preprocessing
Tokenization : simple process converting paragraph into sentence. further sentence --> words--> character depending on case

Terminologies:
Corpus ---> paragraph
Vocabulary ---> unique words
Documents -----> Sentences

```python
corpus = """Hello welcome to nlp world. Please do practical on entire concepts to become expert! do everyday."""
corpus
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
documents
```

```python
type(documents)
```

```python
for sentence in documents:
  print(sentence)
```

```python
#Convert pargraph ---> words or sentence---->words
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
